```python
def register():
    username=input('Create username:')
    password=input('Create password:')
    password1=input('Confirm password:')

    db=open('database.txt','r')
    u=[]
    p=[]
    dbItems = db.read().splitlines()
    
    for i in dbItems:
        a,b=i.split(',')
        u.append(a)
        p.append(b)
    data=dict(zip(u,p))
    db.close()

    if password!=password1:
        print('Password does not match')
        register()
    else:
        if len(password)<=5:
            print('Short password')
            register()
        elif username in u:
            print('Username already exist')
            register()
        else:
            db=open('database.txt','a')
            db.write(username+','+password)
            db.write('\n')
            db.close()
            print('Register sucessfully')

            
def login():
    username=input('Username:')
    password=input('Password:')

    if not len(username or password)<1:
        db=open('database.txt','r')
        u=[]
        p=[]
        dbItems = db.read().splitlines()
        
        for i in dbItems:
            a,b=i.split(',')
            u.append(a)
            p.append(b)
        data=dict(zip(u,p))
        db.close()

        try:
            if username in u:
                try:
                    if data[username] == password:
                        print('Login Successfuly')
                    else:
                        print('Username or password incorrect')
                        
                except:
                    print('incorrect password or username')
            else:
                print('Username not exist')
        except:
            print('Username not exist')
    else:
        print('Please enter the value')

        
def forget():
    db=open('database.txt','r')
    u=[]
    p=[]
    dbItems = db.read().splitlines()
    
    for i in dbItems:
        a,b=i.split(',')
        u.append(a)
        p.append(b)
    data=dict(zip(u,p))
    db.close()
    
    print('CHANGE PASSWORD')
    username=input('Enter the username for the required user:')
    
    try:
        if username in u:
            try:
                oldPassword=data[username]
                newPassword=input('Enter new password')
                oldData=username+','+oldPassword
                newData=username+','+newPassword
                replace(oldPassword, newPassword)
                print('Password updated successfully')
            except:
                print('Error updating new password')
        else:
            print('Username not exist')
    except:
        print('Username not exist')

def replace(oldData, newData):
    file = open("database.txt", "r")
    replacement = ""
    for line in file:
        line = line.strip()
        changes = line.replace(oldData, newData)
        replacement = replacement + changes + '\n'
    file.close()
    
    fout = open("database.txt", "w")
    fout.write(replacement)
    fout.close()
        
        
        
def home(option=None):
    print('Please enter the option')
    option=input('login | signup | forget password\n')
    if option=='login':
        login()
    elif option=='signup':
        register()
    elif option=='forget password':
        forget()
    else:
        home()

home()
```
